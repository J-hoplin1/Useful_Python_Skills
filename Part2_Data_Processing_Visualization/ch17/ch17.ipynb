### 우선 실습 코드내의 디렉토리에 csv파일 목록을 살퍼볼까요?

- os 모듈을 import 해줍니다
- os.getcwd()를 이용하면 현재 디렉토리 정보를 가져올 수 있습니다
- os.listdir을 이용해서 디렉토리 목록을 가져온 후 endswith()를 이용해서 csv확장자만 가져와 봅시다. 이때 filter와 lambda함수를 활용하여 실습합니다
- 이 리스트들을 가지고 사용자가 선택을 하여 읽고자 하는 csv를 선택하게 만들어 봅시다

```python
import os

selected_csv = ""

csv_list = list(filter(lambda x: x.endswith('.csv'),os.listdir(os.getcwd())))

while True:
    print("=" * 15)
    for i,j in enumerate(csv_list,start=1):
        print(f"{i}. {j}")
    print("=" * 15)
    select_number = int(input("읽어오고자 하는 csv파일을 선택해주세요 >> "))
    if select_number < 1 or select_number > len(csv_list):
        print("다시 선택해 주세요")
    else:
        selected_csv = os.getcwd() + "/" + csv_list[select_number - 1]
        break
```

### CSV 파일을 읽어옵니다

- header=None으로 하면 헤더없이 불러올 수 있다는 점을 상기시켜봅시다

```python
df = pd.read_csv(selected_csv)
```

### 선택한 csv파일이 잘 읽어왔는지 확인해볼까요?

```python
print(f"선택한 csv파일명 : {selected_csv.split('/')[-1]}")
df
```

### 앞에서 배운 내용을 상기시켜 봅시다

- index_col을 사용하면, 지정한 index혹은 이름의 열을 기준으로 데이터를 읽어올 수 있었습니다

```python
df2 = pd.read_csv(selected_csv,index_col=0)
```

### 읽어온 csv를 xlsx로 변경해 봅시다

- pd.DataFrame.to_excel()를 사용하면 데이터 프레임을 엑셀로 저장할 수 있습니다.
- to_excel()메소드는 openpyxl라이브러리에 의존성을 지닌 메소드입니다. 위에서 openpyxl을 설치한 이유도 이 이유때문입니다
- 저장할때는 Unnamed 0 열 발생 방지를 위해서 첫번째 행을 인덱스로 읽어들인 데이터 프레임을 통해 저장합니다(df2)

```python
df2.to_excel("csv_to_xls.xlsx")
```

```python
xlsx_df = pd.read_excel("csv_to_xls.xlsx")
```

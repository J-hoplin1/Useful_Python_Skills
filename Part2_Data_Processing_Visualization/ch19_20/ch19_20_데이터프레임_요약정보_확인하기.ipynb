# Ch 19,20 : 데이터프레임의 요약정보 확인하기

### Pandas 활용하기

- Pandas를 import 합니다
- csv파일 하나를 동일 디렉토리로 가져옵니다
- csv파일을 읽어서 DataFrame으로 읽어옵니다

```python
import pandas as pd
import os
```

```python
csv_list = list(filter(lambda x:x.endswith('csv'),os.listdir()))
```

```python
df = pd.read_csv(csv_list[0])
```

```python
df
```

### DataFrame head(), tail() 생각하기

- 앞에서 배운 head(),tail()을 사용해서 각각 7개의 행씩 읽어와 봅니다

```python
df.head(7)
```

```python
df.tail(7)
```

### info()메소드 활용하기

info()메소드를 사용하면, 데이터 프레임의 자료형과 행 인덱스의 종류와 개수, 열의 개수와 자료형, 메모리 총 사용량 등 **데이터프레임의 기본정보**를 확인할 수 있습니다.

```python
df.info()
```

### describe()메소드 활용하기

describe()메소드를 사용하면, **데이터 프레임의 통계정보**를 출력할 수 있습니다. 결과물의 각 필드가 의미하는것은 아래와 같습니다

- count : 각 열의 데이터 개수
- mean : 평균값
- std : 표준편차
- min,max : 최소,최댓값
- 사분위값

```python
df.describe()
```
